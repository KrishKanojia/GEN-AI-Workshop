```python
data = "dataset/spam.csv"
df = pd.read_csv(data)
df
```

# Change Output Feature into Number i.e (0 or 1)

```python
df.groupby('Category').describe()
```

```python
df['spam'] = df['Category'].progress_apply(lambda x : 1 if x == "spam" else 0)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.25, random_state=42)
```

```python
len(X_train.values)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
x_train_count = cv.fit_transform(X_train.values)
```

```python
print(len(cv.vocabulary_))
```

```python
x_train_count.toarray().shape
```

```python
x_train_count.toarray()[0][1756]
```

```python
import numpy as np

# Example: Get the first document's term frequencies as an array
non_zero_indices = np.nonzero(x_train_count.toarray()[0]) # Extract indices

print("Indices with non-zero values:", non_zero_indices)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
x_test_count = cv.transform(X_test)
model.score(x_test_count, y_test)
```

```python
emails = [
    "Alex let's play cricket together",
    "Upto 20%  cashback on your first purchase"
]
emails_count = cv.transform(emails)
model.predict(emails_count)
```

```python
x_train_count.toarray()[0][1756]
```
